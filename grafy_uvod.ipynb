```python
import pylab as pl
from numpy import cos, pi
```

```python
f = 50
t = pl.linspace(0, 0.1, 1000)
u = 3.3 * cos(2*pi*f*t)
i = 1.1 * cos(2*pi*f*t + pi/2)
p = u * i

pl.plot(t, u, "b--", label="u(t)")
pl.plot(t, i, "g:", label="i(t)")
pl.plot(t, p, "r", label="p(t)")
pl.grid()
pl.legend(loc="upper right")
pl.xlim(0, 0.04)
pl.xlabel("t [s]")
pl.ylabel("u(t), i(t), p(t)")
pl.title("Výkon střídavého proudu")
```
