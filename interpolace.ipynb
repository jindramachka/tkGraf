```python
import scipy.interpolate as inp
import pylab as pl

x = [0, 0.3, 0.5, 0.8, 1, 2, 3]
y = [0, 0.1, 0.5, 1, 3, 10, 30]

x_1 = pl.linspace(0, 3, 333)
cubic_spl = inp.CubicSpline(x, y)
univar_spl = inp.UnivariateSpline(x, y)
y_1 = cubic_spl(x_1)
y_2 = univar_spl(x_1)


pl.plot(x, y, "o", label="měřící body")
pl.plot(x_1, y_1, label="kubická interpolace")
pl.plot(x_1, y_2, label="jednorozměrná interpolace")
pl.grid()
pl.legend()
```
